# Prisoners Dilemma

```python
import numpy as np
from IPython import display
from TwoTimesTwoGamesFunctions import updateStrategy, plot, trajectory, plotTrajectory, computeStablePoints, plotStableRegion, checkStability
```

```python
display.Image("PayoffMatrices/PrisonersDilemmaWikipedia.png", width = 300)
```

```python
######### Initialization ######### (note: utility maximization)

# player 1:
payoff1 = np.array([[-1,-3],[0,-2]])

# player 2:
payoff2 = np.array([[-1,0],[-3,-2]])

#number of iterations
iterations = int(200)

#number of vectors in the plot = gridSize^2
gridSize = 15

#game title
gameName = "Prisoner's Dilemma"

#name of first strategy 
nameOfStrategy1 = "Silent"
nameOfStrategy2 = "Silent"

######### Equilibria #########

#Pure Nash Equilibria
pne = [[0,0]]

#Mixed Nash Equilibria 
#sytax: (..., (P1(strategy1),P2(strategy1)), ... )
mne = []
```

```python
########## Online Mirror Ascent One Update Step #############

diff1, diff2 = updateStrategy(iterations, gridSize, payoff1, payoff2, "EGA")
```

```python
#streamPlot:
plot(gridSize, diff1, diff2, pne, mne, gameName, nameOfStrategy1, nameOfStrategy2, "EGA", "stream",
    payoff1, payoff2, [pne[0]])

```

```python
########## Online Mirror Ascent One Update Step #############

diff1, diff2 = updateStrategy(iterations, gridSize, payoff1, payoff2, "POGA")
```

```python
#streamPlot:
plot(gridSize, diff1, diff2, pne, mne, gameName, nameOfStrategy1, nameOfStrategy2, "POGA", "stream",
     payoff1, payoff2, [pne[0]])

```

```python
#plotting a single trajectory

p1Init = 0.8
p2Init = 0.5

p1, p2 = trajectory(p1Init, p2Init, iterations, payoff1, payoff2, "POGA")
plotTrajectory(p1Init, p2Init, p1, p2, pne, mne, gameName, nameOfStrategy1, nameOfStrategy2, "POGA")
```

```python
#plotting a single trajectory

p1Init = 0.8
p2Init = 0.5

p1, p2 = trajectory(p1Init, p2Init, iterations, payoff1, payoff2, "EGA")
plotTrajectory(p1Init, p2Init, p1, p2, pne, mne, gameName, nameOfStrategy1, nameOfStrategy2, "EGA")
```

```python
########## Online Gradient Ascent One Update Step #############

diff1, diff2 = updateStrategy(iterations, gridSize, payoff1, payoff2, "POGA")
```

```python
#streamPlot:
plot(gridSize, diff1, diff2, pne, mne, gameName, nameOfStrategy1, nameOfStrategy2, "POGA", "stream")
```
